```python
#try this function, see which movie is giving you an error
#then take a look at the movie, think about how you would solve it

def get_director(detail_link):  #define your function to get director from detail_link
    detail_page = requests.get(detail_link, headers = headers)
    detail_soup = BeautifulSoup(detail_page.content, "html.parser")
    
    director_table = detail_soup.find('table', class_ = 'credits')
    #print("D",director_table,movie_name)
    if director_table is None:
        return None
    director_line = director_table.find_all('td', class_ = 'person') #there could be more than 1 director
    directors = list()
    for line in director_line:
        print("LINE",line)
        hyperlink = line.find('a')
        director = hyperlink.get_text()
        directors.append(director.strip())
    return directors
```

```python

def get_genres(detail_link):  #define your function to get director from detail_link
    detail_page = requests.get(detail_link, headers = headers)
    detail_soup = BeautifulSoup(detail_page.content, "html.parser")
   
    director_table = detail_soup.find('table', class_ = 'details')
    #print("D",director_table,movie_name)
    if director_table is None:
        return None
    director_line = director_table.find_all('td', class_ = 'data') #there could be more than 1 director
    if director_table is None:
        return None
    director_line = director_table.find_all('tr', class_ = 'genres')
    if director_table is None:
        return None
    #director_line = director_table.find_all('td', class_ = 'data')
    #print(director_line)
    directors = list()
    for line in director_line:
        #print("LINE",line)
        hyperlink = line.find_all('span')
        #print(hyperlink,"HASDF")
        for hyper in hyperlink:
            director = hyper.get_text()
            directors.append(director.strip())
    return directors
```

```python
def movie_detail(movie_name):
    movie = movie_name.title()
    
    idx = df[df.movie_name == movie.title()].index[0]
    
    actor_lis = eval(df.actors[idx])
    actors = ", ".join(actor_lis)
    print("The cast of the movie " + actors + ".")
    
    genre_lis = eval(df.genre[idx])
    genre = ", ".join(genre_lis)
    print("The Genere of the movie is " + genre + ".")
    
```

```python
def people_check(people):
    actor = people.title()
    
    movie_name = []
    
    genre_count = {}
    
    for idx in range(len(df)):
        actor_lis = eval(df.actors[idx])
        
        if actor in actor_lis:
            movie_name.append(df.movie_name[idx])
            
            genre_lis = eval(df.genre[idx])
            
            for gen in genre_lis:
                if gen in genre_count:
                    genre_count[gen] = genre_count[gen]+1
                else:
                    genre_count[gen] = 1
                    
    movies = ", ".join(movie_name)
    
    print(actor + " has acted in " + movies + ".")
    
    
    print("His often played genre are " + ', '.join('{} : {}'.format(key, value) for key, value in genre_count.items()))
    
```

```python
def cosine_similarity(actor1, actor2):
    
    actor1 = actor1.title()
    actor2 = actor2.title()
    
    actor1_genre = {}
    actor2_genre = {}
    
    actor1_movie = []
    actor2_movie = []
    
    for idx in range(len(df)):
        actor_lis = eval(df.actors[idx])
        
        if actor1 in actor_lis:
            actor1_movie.append(df.movie_name[idx])
            genre_lis = eval(df.genre[idx])
            for gen in genre_lis:
                if gen in actor1_genre:
                    actor1_genre[gen] = actor1_genre[gen]+1
                else:
                    actor1_genre[gen] = 1
                
        if actor2 in actor_lis:
            actor2_movie.append(df.movie_name[idx])
            genre_lis = eval(df.genre[idx])
            for gen in genre_lis:
                if gen in actor2_genre:
                    actor2_genre[gen] = actor2_genre[gen]+1
                else:
                    actor2_genre[gen] = 1
    
                
    #print(actor1_genre)
    #print(actor2_genre)
    
    #print(actor1_movie)
    #print(actor2_movie)
    
    set_actor1 = set(actor1_genre)
    set_actor2 = set(actor2_genre)
    combine_set = set_actor1.union(set_actor2)

    #print(combine_set)
    
    actor1_dict = {}
    actor2_dict = {}
    for i in combine_set:
        actor1_dict[i] = 0
        actor2_dict[i] = 0
        
    for idx in range(len(df)):
        actor_lis = eval(df.actors[idx])
        if actor1 in actor_lis:
            genre_lis = eval(df.genre[idx])
            for gen in genre_lis:
                if gen in  actor1_dict:
                    actor1_dict[gen] = actor1_dict[gen]+1
        if actor2 in actor_lis:
            genre_lis = eval(df.genre[idx])
            for gen in genre_lis:
                if gen in  actor2_dict:
                    actor2_dict[gen] = actor2_dict[gen]+1
    #print(actor1_dict)
    #print(actor2_dict)
    
    actor1_vector = list(actor1_dict.values())
    actor2_vector = list(actor2_dict.values())
    #print(actor1_vector)
    #print(actor2_vector)
    
    result = dot(actor1_vector, actor2_vector)/(norm(actor1_vector)*norm(actor2_vector))
    
    
    print(actor1 + " has acted in " + ", ".join(actor1_movie) + ".")
    print("His often played genre are " + ', '.join('{} : {}'.format(key, value) for key, value in actor1_genre.items()))
    
    print(actor2 + " has acted in " + ", ".join(actor2_movie) + ".")
    print("His often played genre are " + ', '.join('{} : {}'.format(key, value) for key, value in actor2_genre.items()))
    
    print("Based on that, they have a cosine similarity score of  "+ str(result) + ".")
    
```
